## 説明変数の選択
### 多重共線性の確認
特徴量間での多重共線性を把握するために、VIF(Variance Inflation Factor)[1]を算出する。<br>
VIFは最小二乗回帰分析における多重共線性の深刻さを定量化できる。<br>
推定された回帰係数の分散(推定値の標準偏差の平方)が、多重共線性のためにどれだけ増加したかを測る指標を提供する。


```python
# 前処理
# 目的変数と説明変数の分割
X = df_diabetes.drop('Outcome', axis=1)
y = df_diabetes['Outcome']

# 説明変数の標準化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# VIFを計算する関数の定義
def calculate_vif(X):
    vif = pd.DataFrame()
    vif["Features"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif

# VIFの計算
vif_df = calculate_vif(pd.DataFrame(X_scaled, columns=X.columns))
print(vif_df)
```

VIFの値が高い場合(10以上)である場合は特徴量かんの相関が高いことを表すが、
算出結果より、すべての特徴量が1に近い値となった。<br>
つまり、特徴量の間で高い相関は観られないことがわかった。

### 交互作用の存在の確認
交互作用の存在を確認するために、Statsmodelsを使ってp値とz値を算出する。<br>
p値,z値はそれぞれ有意性を評価するために用いる指標でp値は0.05以下かつゼロと異なること場合に、有意であることを示す。<br>
また、ｚ値はゼロと異なりかつゼロからの絶対値が大きいほど有意であることを示す。

```python
# 交互作用項を含むモデルの作成
X_interact = sm.add_constant(X)  # 定数項を追加
model = sm.Logit(y, X_interact)  # ロジスティック回帰モデルの作成
result = model.fit()  # モデルの学習

# 結果の表示
result.summary()
```

出力結果より、SkinThickness は p値が 0.929 > 0.05 かつz値が0.090でありゼロに近いため有意性なしと判断する。<br>
Insulin はz値は-1.322でありゼロと1以上離れているが、p値が 0.186 > 0.05 なので有意性なしと判断する。 <br>
Ageはz値は1.593でありゼロと1以上離れているが p値が 0.111 > 0.05 なので有意性なしと判断する <br>
それ以外の特徴量についてはp値が0.05未満もしくはp値が0であってもz値が0より十分大きいため 有意性があると判断する。

## AIC/BICによるモデルのあてはめの評価
前述で選定した特徴量を用いてロジスティック回帰モデルを作成する。<br>
そして、作成したモデルに対して AIC/BICのあてはめ評価を行う。

```python
# 前述で有意性を確認できた特徴量を使って改めてロジスティック回帰モデルを作成
significant_features = ['Pregnancies', 'Glucose', 'BloodPressure', 'BMI', 'DiabetesPedigreeFunction']
significant_features_X = X[significant_features]
scaler = StandardScaler()
significant_features_X_scaled = scaler.fit_transform(significant_features_X)

# ロジスティック回帰モデルの作成
significant_features_X_scaled = sm.add_constant(significant_features_X_scaled)
model = sm.Logit(y, significant_features_X_scaled)
result = model.fit()
```

```python
# モデルのあてはめ評価指標の計算
aic = result.aic
bic = result.bic

print("AIC:", aic)
print("BIC:", bic)
```

出力結果より、目的関数は0.474323で6回のイテレーションで最適化が収束した。<br>
また、AIC/BICの値がそれぞれAIC: 740.5596057697524　BIC: 768.4223441686383となった。<br>
互いが近い値となったためモデルの当てはまりはよいと考える。


## クロスバリデーション、ROC曲線などを用いたモデルの評価

ここではモデルの予測性能を評価する。
- k-forldクロスバリデーションをもちいて平均正解率を算出する。<br>
- ROC曲線をプロットし、予測性能を可視化する。<br>
- 混同行列とF値を算出しモデルの性能を評価する。<br>


※ ここまで使用してきた statsmodels にはクロスバリデーション機能が組み込まれていない。
また、Scikit-learnとはインタフェースが異なるため、scikit-learnでモデルを作り直す。


```python
X = df_diabetes.drop('Outcome', axis=1)
y = df_diabetes['Outcome'].astype(int)

significant_features = ['Pregnancies', 'Glucose', 'BloodPressure', 'BMI', 'DiabetesPedigreeFunction']
significant_features_X = X[significant_features]

scaler = StandardScaler()
significant_features_X_scaled = scaler.fit_transform(significant_features_X)

# 訓練データとテストデータの分割
X_train, X_test, y_train, y_test = train_test_split(significant_features_X_scaled, y, test_size=0.2, random_state=42)

# ロジスティック回帰モデルのインスタンス化と訓練
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# クロスバリデーションによるモデルの評価
scores = cross_val_score(logreg, X_train, y_train, cv=5)  # 5-foldクロスバリデーション
mean_accuracy = scores.mean()  # 平均正解率の計算
print("平均正解率:", mean_accuracy)

# 予測クラスを取得
y_pred = logreg.predict(X_test)

# テストデータを用いた予測確率の取得
y_pred_proba = logreg.predict_proba(X_test)[:, 1]

# F値の計算
f_score = f1_score(y_test, y_pred)
print('F score:', f_score)

# 混同行列の計算
confusion = confusion_matrix(y_test, y_pred)
print(confusion)

# 混同行列の可視化
sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

# ROC曲線の計算
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# ROC曲線のプロット
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()

# AUCスコアの計算
auc_score = roc_auc_score(y_test, y_pred_proba)
```

算出結果より、平均正解率約77.4%で予測可能とわかった。<br>
F値についても0.66 となり、0から1の尺度では 1に近く一定の予測水準は保持していると示唆する。<br>
また、ROC曲線も一般的によくみられるカーブを描いており、不安定な形とはなっていない。<br>
総じて、高いモデル性能ではないものの、疾患予測の１つの参考程度として用いることができる性能はもたせることはできたと示唆する。
